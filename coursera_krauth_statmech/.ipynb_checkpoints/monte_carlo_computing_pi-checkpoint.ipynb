```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def compute_pi(nrep, ns):
    samples = 2.0 * (np.random.rand(nrep, 2, ns) - 0.5)
    dist = np.sqrt((samples**2).sum(axis=1))
    ratio = (dist <= 1.0).sum(axis=1)/ns
    pi_comp = 4 * ratio
    return pi_comp
```

```python
plt.rcParams['figure.figsize'] = (12.0, 10.0)
nsmp = 123
nreps = np.power(3, np.arange(3, 12))
rng = (2.74, 3.54)
mus = np.zeros((1, nrep.shape[0]))
sts = np.zeros((1, nrep.shape[0]))
for i, nrep in enumerate(nreps):
    pi_comp = compute_pi(nrep, nsmp)
    mus[i] = pi_comp.mean()
    sts[i] = pi_comp.std()
    plt.subplot(3, 3, i + 1)
    a = plt.hist(pi_comp, bins = 50, range=rng, normed=False)
    plt.xlim(rng)
    plt.title('mean = {:1.4f} std = {:1.4f}'.format(mu, st))
```
